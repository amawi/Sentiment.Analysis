## In this notebook, I will calculate the average sentiment over all the 7 Weeks/ Chapters.

```python
import numpy as np
```

```python
# We save all 7 chapters vars and their values.
week1score = 0.95343125
week2score = 0.29841249999999997
week3score = 0.37424999999999997
week4score = 0.48164375
week5score = 0.28245624999999996
week6score = 0.40803125
week7score = 0.38788125
```

```python
# We make a list so we can use np average
week_scores_list = [week1score, week2score, week3score, week4score, week5score,week6score, week7score]
```

```python
#We calculate the average using numpy
average = np.mean(week_scores_list)
print("The average of the scores for the 7 weeks or chapters is:",  average)
```

```python
import matplotlib.pyplot as plt

# Sample list of scores over time
scores = [week1score, week2score, week3score, week4score, week5score,week6score, week7score]

# Create a line plot to visualize the changes in scores
plt.figure(figsize=(10, 6))
plt.plot(scores, marker='o', linestyle='-', color='b')
plt.title('Change in Scores Each Week')
plt.xlabel('Chapters')
plt.ylabel('Sentiment Scores')
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample list of scores over time
scores = [week1score, week2score, week3score, week4score, week5score,week6score, week7score]

# Set Seaborn style
sns.set(style="whitegrid")

# Create a line plot to visualize the changes in scores
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(len(scores)), y=scores, marker='o', color='b')
plt.title('Change in Scores Over Time')
plt.xlabel('Time Period')
plt.ylabel('Score')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample list of scores over time
scores = [week1score, week2score, week3score, week4score, week5score,week6score, week7score]

# Corresponding time periods
time_periods = ['Week1', 'Week2', 'Week3', 'Week4', 'Week5', 'Week6', 'Week7']

# Create a bar plot to visualize the changes in scores
plt.figure(figsize=(12, 8))
bars = plt.bar(time_periods, scores, color='skyblue')
plt.title('Change in Scores Over Time')
plt.xlabel('Time Period')
plt.ylabel('Score')

# Add scores inside the upper side of the bars with 2 decimal places
for bar, score in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() - 0.02, f'{score:.3f}', fontsize=12, ha='center')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
